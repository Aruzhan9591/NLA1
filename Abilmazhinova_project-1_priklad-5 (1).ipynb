```python
import numpy as np

# Definujeme A, b
A = np.array([
    [4, 1, 0],
    [1, 3, 1],
    [0, 1, 2]
], dtype=float)

b = np.array([1, 2, 0], dtype=float)
x0 = np.zeros(3)

# Iterace 0
r0 = b - A @ x0
d0 = r0.copy()
delta_0 = r0 @ r0

# Iterace 1
Ad0 = A @ d0
alpha0 = delta_0 / (d0 @ Ad0)
x1 = x0 + alpha0 * d0
r1 = r0 - alpha0 * Ad0
delta_1 = r1 @ r1
beta0 = delta_1 / delta_0
d1 = r1 + beta0 * d0

# Iterace 2
Ad1 = A @ d1
alpha1 = delta_1 / (d1 @ Ad1)
x2 = x1 + alpha1 * d1
r2 = r1 - alpha1 * Ad1

# Přesné řešení pomocí knihovní funkce
x_star = np.linalg.solve(A, b)
error = np.linalg.norm(x2 - x_star)

print("x0 =", x0)
print("x1 =", x1)
print("x2 =", x2)
print("Přesné řešení x* =", x_star)
print("Rozdíl ||x2 - x*|| =", error)
```
