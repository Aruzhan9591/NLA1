```python
import numpy as np

def householder_matrix(x, y):
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)

    # Normalizace směrového vektoru y
    y_norm = y / np.linalg.norm(y)

    # Výpočet vektoru v
    alpha = np.linalg.norm(x)
    v = x - alpha * y_norm

    if np.allclose(v, 0):
        return np.identity(len(x))

    v = v.reshape(-1, 1)
    P = np.identity(len(x)) - 2 * (v @ v.T) / (v.T @ v)
    return P
```

```python
x = [1, 2, 3, 4]
y = [4, 3, 2, 1]

P = householder_matrix(x, y)

Px = P @ x
print("P @ x =", Px)

# Ověření směru
Px_norm = Px / np.linalg.norm(Px)
y_norm = np.array(y) / np.linalg.norm(y)

print("Normalizované Px:", Px_norm)
print("Normalizované y:", y_norm)

# cos_angle = np.dot(Px_norm, y_norm)
# print("cos(úhlu) mezi Px a y =", cos_angle)
```

```python
# Druhý náhodný příklad
x2 = [5, -2, 1, 0]
y2 = [-1, 2, 4, -3]

# Výpočet Householderovy matice pro nové vektory
P2 = householder_matrix(x2, y2)

# Výpočet transformovaného vektoru
Px2 = P2 @ x2
print("P2 @ x2 =", Px2)

# Normalizace vektorů pro porovnání směru
Px2_norm = Px2 / np.linalg.norm(Px2)
y2_norm = np.array(y2) / np.linalg.norm(y2)

print("Normalizované P2x:", Px2_norm)
print("Normalizované y2:", y2_norm)

# # Výpočet kosinu úhlu mezi výsledkem a cílovým vektorem
# cos_angle2 = np.dot(Px2_norm, y2_norm)
# print("cos(úhlu) mezi P2x a y2 =", cos_angle2)
```
